Exercises #1-5 use data collected on American hospitals that are registered with Medicare, a federal health insurance program for those 65 or older, or younger patients with certain medical conditions. This data is published online to allow patients to make more informed decisions.

**Run the following cell before continuing to create the DataFrame, `hospital_df`.**

```python
url = "https://raw.githubusercontent.com/the-codingschool/datascience/master/Unit%203/Outpatient_Procedures.csv"
hospital_df = pd.read_csv(url)
hospital_df
```

#### **Exercise #1:** How many null values does the column `eye` have?

```python
hospital_df.info()

#4163 total entries - 3448 non-null = 715 nulls
```

#### **Exercise #2:** Delete the column `Unnamed: 0`.

```python
hospital_df = hospital_df.drop(columns = "Unnamed: 0")
```

#### **Exercise #3:** Create a new DataFrame and change any null values in the DataFrame to 0.0. What is the new 4158th value in the `nervous_system` column *after the null values have been replaced*?

```python
new_df1 = hospital_df.fillna(value = 0.0)
new_df1["other"][4159]
```

#### **Exercise #4:** Using `hospital_df` again, which still has null values, create a new column called `out` that is the sum of the `gastrointestinal`, `eye`, `nervous_system`, `musculoskeletal`, `skin`, `genitourinary`, `cardoivascular`, and `respiratory` columns. What is the 1033rd value in this new column?

```python
out = hospital_df["gastrointestinal"] + hospital_df['eye'] + hospital_df['nervous_system'] + hospital_df['musculoskeletal'] + hospital_df['skin'] + hospital_df['genitourinary'] + hospital_df['cardiovascular'] + hospital_df['respiratory']
hospital_df['out'] = out
print(hospital_df['out'][1032])
```

#### **Exercise #5:** Sort the dataframe on the `out` column in descending order. What is the 100th value in the `musculoskeletal` column of this sorted dataframe? *You cannot use the index to get this value, so what should you do instead?*

```python
hospital_df = hospital_df.sort_values('out', ascending=False)
hospital_df.head(100)
```

## Part 2: Weston, MA Dataset
---
The following dataset shows information about houses for sale in Weston, MA. You will perform several data wrangling/cleaning techniques on this dataset.

**Run the following cell before continuing to create the DataFrame, `houses_df`.**

```python
house_dict = {"town":["Weston", "Weston", "Weston", "Weston", "Weston", "Weston", "Weston", "Weston", "Weston"],
              "state": ["MA", "MA", "MA", "MA", "MA", "MA", "MA", "MA", "MA"],
              "street": ["Webster", "Perry", "Loring", "Wellsley", "Lantern", "Jericho", "Lexington", "Colchester", "Partridge"],
              "price": [10400000, 9750000, None, None, 1790000, 799000, 1995000, 4250000, 2195000],
              "feet": [14056, 12795, None, None, 4663, 1611, 4600, 6700, 4070]
    
}
houses_df = pd.DataFrame(house_dict)
houses_df
```

#### **Exercise #6:** Using the `houses_df` dataframe provided, fill in the NaN values in the `price` column with the average of that column. What is this average?

```python
from pandas._libs.hashtable import value_count
avg = houses_df['price'].mean()
houses_df['price'] = houses_df['price'].fillna(value = avg)
print(avg)
```

#### **Exercise #7:** Fill in the NaN values in the `feet` column with the average of that column. What is this average?

```python
feet_avg = houses_df['feet'].mean()
houses_df = houses_df['feet'].fillna(value = feet_avg)
print(feet_avg)
```

#### **Exercise #8:** Create a new column `price_per_sq_ft` that has the price per square foot for each house. What is the 5th value in this new column?

```python
price_per_sq_ft = houses_df['price'] / houses_df['feet']
houses_df['price_per_sq_ft'] = price_per_sq_ft
print(houses_df.head())
```
