### **Exercise #2**

Initialize and train 4 Decision trees such that:
* All features are used as inputs
* The "winner" column is used as the target/class
* One model has no restrictions on its depth (or anything else)
* One model has a maximum depth of 8
* One model has a maximum depth of 3
* One model has a maximum depth of your choice (feel free to play with this value)
* All models have a different, meaningful variable name (ex: `tree_1` or `depth3_tree` instead of just `model`)

<br>


**NOTE**: Remember the cheat sheet is provided above for you to use.

```python
inputs = df.drop(columns = ['winner'], axis = 1)
output = df['winner']

X_train, X_test, y_train, y_test = train_test_split(inputs, output, test_size = 0.2, random_state = 42)
tree1 = DecisionTreeClassifier()
tree1.fit(X_train, y_train)
```

```python
inputs = df.drop(columns = ['winner'], axis = 1)
output = df['winner']

X_train, X_test, y_train, y_test = train_test_split(inputs, output, test_size = 0.2, random_state = 42)
tree2 = DecisionTreeClassifier(max_depth=8)
tree2.fit(X_train, y_train)
```

```python
inputs = df.drop(columns = ['winner'], axis = 1)
output = df['winner']

X_train, X_test, y_train, y_test = train_test_split(inputs, output, test_size = 0.2, random_state = 42)
tree3 = DecisionTreeClassifier(max_depth=3)
tree3.fit(X_train, y_train)
```

```python
inputs = df.drop(columns = ['winner'], axis = 1)
output = df['winner']

X_train, X_test, y_train, y_test = train_test_split(inputs, output, test_size = 0.2, random_state = 42)
tree4 = DecisionTreeClassifier(max_depth=5)
tree4.fit(X_train, y_train)
```

### **Exercise #3**

Evalute all 4 models using the confusion matrix and classification report and determine which one is best.

**1. Model 1 with no restrictions on the depth**

```python
predictions1 = tree1.predict(X_test)
plot_confusion_matrix(tree1, X_test, y_test)

print("Report: " + str(classification_report(y_test, predictions1)))
```

**2. Model 2 with a max depth of 8**

```python
predictions2 = tree2.predict(X_test)
plot_confusion_matrix(tree2, X_test, y_test)

print("Report: " + str(classification_report(y_test, predictions2)))
```

**3. Model 3 with a max depth of 3**

```python
predictions3 = tree3.predict(X_test)
plot_confusion_matrix(tree3, X_test, y_test)

print("Report: " + str(classification_report(y_test, predictions3)))
```

**4. Model 4 with a max depth of your choice**

```python
predictions4 = tree4.predict(X_test)
plot_confusion_matrix(tree4, X_test, y_test)

print("Report: " + str(classification_report(y_test, predictions4)))
```

### **Exercise #4**

### **Exercise #6**
---

Visualize the feature importances according to the unrestricted tree. Since these features represent locations on the board, we can use this to formulate a potential strategy for playing tic-tac-toe.

```python
plt.figure(figsize = (8, 5))
plt.bar(X_test.columns, tree1.feature_importances_)

plt.title("Importance of Each Feature According to Decision Tree", fontsize = "x-large")
plt.xlabel("Feature", fontsize = "x-large")
plt.ylabel("Importance", fontsize = "x-large")

plt.xticks(rotation = 90)

plt.show()
```
